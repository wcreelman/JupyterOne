```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

#parameters
h = 0.0       #starting height
a = -9.81     #acceleration due to gravity
angle = 45.0  #degrees upward
v = 20.0      #m/s
t = 0.0       #time at start of launch
deltaT = 0.1  #time intervals to plot
theta = angle * math.pi / 180.0
x = [0]
y = [h]

#graphing code
fig, ax = plt.subplots()
ax.set_xlim(0, 100)    #these numbes presumably change with each problem, need to calculate
ax.set_ylim(0, 150)
motion, = ax.plot([],[
    
])

def animation_frame(t):  
    dx = v * math.cos(theta) * t
    vy = v * math.sin(theta) + a * t
    h = y[-1] +  (vy * deltaT)
    print(t, dx, h)  
    motion.set_data(dx, h)
    return motion,
    
animation = FuncAnimation(fig, func=animation_frame, frames=np.arange(0, 100, 10), interval=200)

"""
"""    
plt.xlabel('X')
plt.ylabel('Height')
plt.title("Projectile Motion")

#plt.scatter(x, y, animation_frame="time")
plt.show()
```
